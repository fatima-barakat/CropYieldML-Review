Clean column names

```python
def clean(df):
    df.columns = (
        df.columns.str.strip()
                  .str.lower()
                  .str.replace(" ", "_")
    )
    return df

df_yield      = clean(df_yield)
df_rainfall   = clean(df_rainfall)
df_temp       = clean(df_temp)
df_pesticides = clean(df_pesticides)
```

Harmonize temperature (country → area)

```python
df_temp = df_temp.rename(columns={"country": "area"})
```

Rename ambiguous value columns

```python
df_yield = df_yield.rename(columns={"value": "yield_value"})
df_pesticides = df_pesticides.rename(columns={"value": "pesticide_value"})
```

Perform the correct merge

```python
df_merged = (
    df_yield
    # Climate (area/year)
    .merge(df_rainfall, on=["area", "year"], how="left")
    .merge(df_temp, on=["area", "year"], how="left")

    # Pesticides (area/year/item)
    .merge(
        df_pesticides[["area", "year", "item", "pesticide_value"]],
        on=["area", "year", "item"],
        how="left"
    )
)

print(df_merged.info())
```

Convert numeric columns explicitly

```python
# Convert climate values to float
df_rainfall["average_rain_fall_mm_per_year"] = pd.to_numeric(
    df_rainfall["average_rain_fall_mm_per_year"], errors="coerce"
)
df_temp["avg_temp"] = pd.to_numeric(df_temp["avg_temp"], errors="coerce")
```

Aggregate after conversion

```python
df_rainfall_agg = df_rainfall.groupby(["area","year"])["average_rain_fall_mm_per_year"].mean().reset_index()
df_temp_agg     = df_temp.groupby(["area","year"])["avg_temp"].mean().reset_index()
```

Merge with yield & pesticides

```python
df_merged = (
    df_yield
    .merge(df_rainfall_agg, on=["area", "year"], how="left")
    .merge(df_temp_agg, on=["area", "year"], how="left")
    .merge(
        df_pesticides[["area","year","item","pesticide_value"]],
        on=["area","year","item"],
        how="left"
    )
)

print("Duplicates:", df_merged.duplicated(subset=["area","year","item"]).sum())
```

##Feature Analysis

```python
# ===============================
# 01_exploration - Feature Analysis
# ===============================

import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# ---- Step 1: Identify numeric vs categorical ----
numeric_cols = df_merged.select_dtypes(include=['int64', 'float64']).columns.tolist()
categorical_cols = df_merged.select_dtypes(include=['object']).columns.tolist()

print("Numeric columns:", numeric_cols)
print("Categorical columns:", categorical_cols)

# ---- Step 2: Convert possible numeric strings to numeric ----
# Example: climate and pesticide columns
for col in ["avg_temp", "average_rain_fall_mm_per_year", "pesticide_value"]:
    if col in df_merged.columns:
        df_merged[col] = pd.to_numeric(df_merged[col], errors='coerce')

# Update numeric columns after conversion
numeric_cols = df_merged.select_dtypes(include=['int64', 'float64']).columns.tolist()

# ---- Step 3: Correlation with target ----
correlation = df_merged[numeric_cols].corr()["yield_value"].sort_values(ascending=False)
print("Correlation with yield_value:")
print(correlation)

# ---- Step 4: Distribution plots for numeric features ----
for col in numeric_cols:
    if col != "yield_value":
        plt.figure(figsize=(6,4))
        sns.scatterplot(data=df_merged, x=col, y="yield_value", alpha=0.6)
        plt.title(f"Yield vs {col}")
        plt.show()

# ---- Step 5: Boxplots / value counts for categorical features ----
for col in categorical_cols:
    plt.figure(figsize=(8,4))
    if df_merged[col].nunique() < 20:
        sns.boxplot(data=df_merged, x=col, y="yield_value")
        plt.title(f"Yield distribution by {col}")
        plt.show()
    else:
        print(f"Top categories in {col}:", df_merged[col].value_counts().head(10))

# ---- Step 6: Optional - Encode categorical for analysis ----
# Convert categorical columns to numeric codes for quick correlation / heatmaps
for col in categorical_cols:
    df_merged[col + "_code"] = df_merged[col].astype("category").cat.codes

# Updated correlation including encoded categories
numeric_cols = df_merged.select_dtypes(include=['int64', 'float64']).columns.tolist()
correlation = df_merged[numeric_cols].corr()["yield_value"].sort_values(ascending=False)
print("Correlation with yield_value including encoded categories:")
print(correlation)
```
